```python
import argparse
import os
import random
import torch
from torch import nn
from torchvision import transforms
from transformers import CLIPModel, CLIPTokenizer
import pandas as pd
from PIL import Image
```

```python
def load_dataset(df):
    """
    Load DataFrame from pickle. Assumes columns: 'pil_image' (PIL.Image), 'combined_text' (str).
    Returns lists of PIL images and text strings.
    """

    images = df['pil_image'].tolist()
    texts = df['combined_text'].tolist()
    return images, texts
```

```python
def compute_image_text_contrastive_loss(image_embeds, text_embeds, temperature=0.07):
    """
    Symmetric cross-modal InfoNCE loss between image and text embeddings.
    """
    image_embeds = nn.functional.normalize(image_embeds, dim=1)
    text_embeds = nn.functional.normalize(text_embeds, dim=1)
    logits = torch.matmul(image_embeds, text_embeds.T) / temperature
    labels = torch.arange(image_embeds.size(0), device=image_embeds.device)
    loss_i2t = nn.functional.cross_entropy(logits, labels)
    loss_t2i = nn.functional.cross_entropy(logits.T, labels)
    return 0.5 * (loss_i2t + loss_t2i)

```

```python
def get_image_transform():
    return transforms.Compose([
        # If it’s not already a PIL Image, convert from array; otherwise, just ensure RGB.
        transforms.Lambda(lambda img: img.convert("RGB")
                                  if isinstance(img, Image.Image)
                                  else Image.fromarray(img).convert("RGB")),
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=(0.48145466, 0.4578275, 0.40821073),
            std=(0.26862954, 0.26130258, 0.27577711),
        ),
    ])
```

```python
def train(pkl_path,
          output_dir,
          batch_size,
          lr,
          temperature,
          device,
          epochs=50,
          patience=3):
    # Load data
    print(f"Loading dataset from {pkl_path}…")
    images, texts = load_dataset(df_combined)
    assert len(images) == len(texts), "Image/Text count mismatch"
    num_samples = len(images)
    print(f"Loaded {num_samples} samples.")

    # Prepare CLIP
    print("Initializing CLIP model and tokenizer…")
    model = CLIPModel.from_pretrained('openai/clip-vit-base-patch32').to(device)
    tokenizer = CLIPTokenizer.from_pretrained('openai/clip-vit-base-patch32')
    model.train()

    img_transform = get_image_transform()
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr)
    os.makedirs(output_dir, exist_ok=True)

    # Early-stopping state
    best_loss = float('inf')
    epochs_no_improve = 0
    best_ckpt_path = os.path.join(output_dir, "best_clip_finetune.pt")

    for epoch in range(1, epochs + 1):
        print(f"\nEpoch {epoch}/{epochs}")
        # shuffle
        idxs = list(range(num_samples))
        random.shuffle(idxs)
        total_loss = 0.0
        steps = 0

        for start in range(0, num_samples, batch_size):
            batch_idxs = idxs[start:start + batch_size]
            if len(batch_idxs) < batch_size:
                break

            # images → tensor
            imgs = [img_transform(images[i]) for i in batch_idxs]
            pixel_values = torch.stack(imgs).to(device)

            # texts → token ids
            batch_texts = [texts[i] for i in batch_idxs]
            inputs = tokenizer(
                batch_texts,
                padding=True,
                truncation=True,
                return_tensors='pt'
            ).to(device)
            print("\n[Debug] Batch inputs:")
            print("Texts:")
            for text in batch_texts:
                print(f"  • {text}")

            print(f"Pixel values tensor shape: {pixel_values.shape}")
            print(f"Input IDs shape: {inputs.input_ids.shape}")
            print(f"Attention mask shape: {inputs.attention_mask.shape}")
            # forward
            outputs = model(
                pixel_values=pixel_values,
                input_ids=inputs.input_ids,
                attention_mask=inputs.attention_mask,
                return_loss=False,
                output_attentions=False,
            )
            image_embeds = outputs.image_embeds
            text_embeds  = outputs.text_embeds

            # contrastive loss
            loss = compute_image_text_contrastive_loss(
                image_embeds, text_embeds, temperature
            )
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            steps += 1
            if steps % 10 == 0:
                print(f"  Step {steps} – loss {loss.item():.4f}")

        avg_loss = total_loss / steps if steps else float('inf')
        print(f"Epoch {epoch} done. Avg loss {avg_loss:.4f}")

        # early-stop check
        if avg_loss < best_loss:
            best_loss = avg_loss
            epochs_no_improve = 0
            torch.save(model.state_dict(), best_ckpt_path)
            print(f"  🎉 New best model (loss {best_loss:.4f}) saved.")
        else:
            epochs_no_improve += 1
            print(f"  No improvement for {epochs_no_improve} epoch(s).")

        if epochs_no_improve >= patience:
            print(f"\n🏁 Early stopping: no improvement in {patience} epochs.")
            break

        # optionally still save per-epoch
        epoch_ckpt = os.path.join(
            output_dir, f"clip_finetune_epoch{epoch}.pt"
        )
        torch.save(model.state_dict(), epoch_ckpt)

    print("\nTraining complete.")
    print(f"Best model at {best_ckpt_path} (loss {best_loss:.4f}).")
```

```python
import torch
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
output_dir = "/content/drive/MyDrive/sbert_histopathology_minilm"
# Helper to move batch to device
def batch_to_device(batch, target_device):
    """
    Send a batch to a device
    """
    for key in batch:
        if isinstance(batch[key], torch.Tensor):
            batch[key] = batch[key].to(target_device)
    return batch

# 1) Simple dataset for raw text
class CaptionDataset(Dataset):
    def __init__(self, captions):
        self.captions = captions

    def __len__(self):
        return len(self.captions)

    def __getitem__(self, idx):
        return self.captions[idx]

# Assuming df is loaded from somewhere
# df = pd.read_csv("your_captions_file.csv")

# 2) Initialize MiniLM
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name, device=device)

# Set dropout for SimCSE approach
for module in model.modules():
    if isinstance(module, torch.nn.Dropout):
        module.p = 0.1  # Set dropout probability for SimCSE

# 3) Prepare DataLoader with raw text data
captions = df["combined_text"].tolist()
dataset = CaptionDataset(captions)
batch_size = 32

loader = DataLoader(
    dataset,
    batch_size=batch_size,
    shuffle=True,
    num_workers=2
)

# 4) Loss & optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
temperature = 0.05  # Temperature parameter for InfoNCE loss

# 5) SimCSE-style training
model.train()
num_epochs = 50  # Increased to max 50 epochs
patience = 3     # Early stopping patience
best_loss = float('inf')
patience_counter = 0

for epoch in range(num_epochs):
    total_loss = 0
    steps = 0

    # Add progress bar for better monitoring
    progress_bar = tqdm(loader, desc=f"Epoch {epoch+1}/{num_epochs}")

    for batch in progress_bar:
        # Make sure batch is a list of sentences
        if isinstance(batch, torch.Tensor):
            batch = batch.tolist()

        # Handle various input types
        sentences = []
        for item in batch:
            if isinstance(item, str):
                sentences.append(item)
            elif isinstance(item, list) and all(isinstance(s, str) for s in item):
                sentences.extend(item)
            elif hasattr(item, 'item'):  # Tensor with single value
                sentences.append(str(item.item()))
            else:
                sentences.append(str(item))

        print("Sentences:")
        for sent_idx, sentence in enumerate(sentences, 1):
            print(f"  {sent_idx}. {sentence}")
        print(f"Total sentences in batch: {len(sentences)}")
        # Get token features from model's tokenizer
        features = model.tokenize(sentences)
        features = batch_to_device(features, device)

        # First encoding
        out_features = model(features)
        embeddings_a = out_features['sentence_embedding']

        # Second encoding with same inputs (SimCSE approach - dropout creates differences)
        out_features = model(features)
        embeddings_b = out_features['sentence_embedding']

        # Normalize embeddings for cosine similarity
        embeddings_a = torch.nn.functional.normalize(embeddings_a, p=2, dim=1)
        embeddings_b = torch.nn.functional.normalize(embeddings_b, p=2, dim=1)

        # Calculate cosine similarity
        sim_matrix = torch.matmul(embeddings_a, embeddings_b.transpose(0, 1)) / temperature

        # SimCSE loss = InfoNCE loss with positives on diagonal
        labels = torch.arange(sim_matrix.size(0), device=device)
        loss = torch.nn.CrossEntropyLoss()(sim_matrix, labels)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Update metrics
        total_loss += loss.item()
        steps += 1

        # Update progress bar
        progress_bar.set_postfix({'loss': loss.item()})

    # Print epoch statistics
    avg_loss = total_loss / steps
    print(f"Epoch {epoch+1}/{num_epochs}: Average loss = {avg_loss:.4f}")

    # Early stopping check
    if avg_loss < best_loss:
        best_loss = avg_loss
        patience_counter = 0
        print(f"New best loss: {best_loss:.4f} - Saving model")

        # Save best model
        best_model_dir = f"{output_dir}_best"
        model.save(best_model_dir)
        print(f"✅ Saved best model to {best_model_dir}")
    else:
        patience_counter += 1
        print(f"No improvement for {patience_counter} epochs. Best loss: {best_loss:.4f}")

        if patience_counter >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break

# 6) Save final model

model.save(output_dir)
print(f"✅ Saved final model to {output_dir}")

print("Training completed!")
print(f"Best loss: {best_loss:.4f}")
print(f"Best model saved to: {output_dir}_best")
```
