```python
import os
import pandas as pd
from PIL import Image
from torch.utils.data import Dataset
import glob

class GastricDataset(Dataset):
    def __init__(self, root_dir, captions_file, processor, image_exts=("jpg", "png", "jpeg"), log_every=10000):
        self.root_dir = root_dir
        self.processor = processor
        self.data = []

        print(f"Reading caption file: {captions_file}")
        df = pd.read_csv(captions_file)
        id_to_caption = {str(row["id"]): row["text"] for _, row in df.iterrows()}
        print(f"Total captions loaded: {len(id_to_caption)}")

        print(f"Scanning image directory (glob-based): {root_dir}")
        match_count = 0
        scanned = 0
        pattern = os.path.join(root_dir, '**', '*.*')

        for filepath in glob.iglob(pattern, recursive=True):
            scanned += 1
            filename = os.path.basename(filepath)

            if not filename.lower().endswith(image_exts):
                continue
            if "_" not in filename:
                continue

            scan_id = filename.split("_")[0]
            caption = id_to_caption.get(scan_id)
            if caption:
                self.data.append({"image_path": filepath, "caption": caption})
                match_count += 1

            if scanned % log_every == 0:
                print(f"Scanned {scanned} files... Matched pairs: {match_count}")

        if not self.data:
            raise ValueError("No valid image-caption pairs found.")

        print(f"Completed. Total image-caption pairs: {len(self.data)}")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        image = Image.open(item["image_path"]).convert("RGB")

        inputs = self.processor(
            images=image,
            text=item["caption"],
            return_tensors="pt",
            do_rescale=False
        )
        return {k: v.squeeze(0) for k, v in inputs.items()}


# Instantiate dataset
train_dataset = GastricDataset(
    root_dir="/content/drive/MyDrive/patches_captions/patches_captions",
    captions_file="/content/drive/MyDrive/patches_captions/captions.csv",
    processor=processor
)

```

```python
# Custom collate function
def collate_fn(batch):
    return processor.pad(batch, return_tensors="pt")

# Training arguments
training_args = TrainingArguments(
    output_dir="./blip2_gastric_finetuned",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    save_strategy="epoch",
    logging_dir="./logs",
    fp16=True,
    remove_unused_columns=False,
    report_to="none"  # to avoid WANDB warnings
)

# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    tokenizer=processor,
    data_collator=collate_fn,
)

# Train
trainer.train()

# Save final model
model.save_pretrained("./blip2_gastric_finetuned")
processor.save_pretrained("./blip2_gastric_finetuned")
```

```python
import os
import pandas as pd
from PIL import Image

def process_images_from_captions_direct(root_dir, captions_file, log_every=10):
    df = pd.read_csv(captions_file)
    data = []

    for i, row in df.iterrows():
        scan_id = str(row["id"])
        caption = row["text"]
        print(f"Processing scan_id: {scan_id}")
        # Match files starting with scan_id and ending in .jpg (case-insensitive)
        possible_files = [
            f for f in os.listdir(root_dir)
            if f.lower().endswith(".jpg") and f.startswith(f"{scan_id}_")
        ]

        for file_name in possible_files:
            file_path = os.path.join(root_dir, file_name)
            try:
                image = Image.open(file_path).convert("RGB")
                data.append({
                    "filename": file_name,
                    "image": image,
                    "caption": caption
                })
            except Exception as e:
                print(f"âŒ Error reading {file_name}: {e}")

        if (i + 1) % log_every == 0:
            print(f"âœ… Processed {i+1}/{len(df)} caption rows. Pairs so far: {len(data)}")

    return pd.DataFrame(data)
```

```python
import pandas as pd
from google.cloud import storage
from io import BytesIO
from PIL import Image
from tqdm import tqdm

# === Step 1: Load CSV ===
captions_df = pd.read_csv("/content/drive/MyDrive/patches_captions/captions.csv")
captions_df['id'] = captions_df['id'].astype(str)
print(f"Loaded {len(captions_df)} rows from captions.csv")

# === Step 2: Setup GCS Client ===
client = storage.Client(project="elated-coil-451021-r0")
bucket = client.get_bucket("cancerhistpath_abdcancer")

# === Step 3: List blobs and index ALL that match each ID ===
print("Fetching file list from GCS...")
blobs = list(bucket.list_blobs(prefix="dataset-upload_cancerhistnew/"))

# Change: store list of blobs per file_id
blob_dict = {}
for blob in blobs:
    filename = blob.name.split("/")[-1]
    if "_" in filename:
        file_id = filename.split("_")[0]
        blob_dict.setdefault(file_id, []).append(blob)

total_matched_images = sum(len(v) for v in blob_dict.values())
print(f"Indexed {total_matched_images} image files grouped by ID from GCS")

# === Step 4: Process and allow multiple images per ID ===
records = []
skipped = 0
failed_images = 0

print("Processing rows and matching with images...")

for _, row in tqdm(captions_df.iterrows(), total=len(captions_df)):
    file_id = row['id']

    if file_id not in blob_dict:
        skipped += 1
        continue

    for blob in blob_dict[file_id]:
        try:
            image_bytes = blob.download_as_bytes()
            pil_image = Image.open(BytesIO(image_bytes)).convert("RGB")
        except Exception as e:
            print(f"Failed to load image for ID {file_id}: {e}")
            failed_images += 1
            continue

        combined_text = f"{row['subtype']} - {row['text']}"
        records.append({
            "file_name": blob.name.split("/")[-1],
            "pil_image": pil_image,
            "combined_text": combined_text
        })

# === Step 5: Save Final Dataset ===
output_df = pd.DataFrame(records)
output_df.to_pickle("histopath_dataset_correct.pkl")

print(f"\nâœ… Saved dataset with {len(output_df)} image-caption entries.")
print(f"âŒ Skipped {skipped} captions (no image match).")
print(f"âš ï¸ Failed to load {failed_images} images.")
```

```python
embedded_records = []

for _, row in tqdm(df.iterrows(), total=len(df)):
    try:
        # Prepare image
        inputs = clip_processor(images=row["pil_image"], return_tensors="pt").to(device)
        with torch.no_grad():
            img_emb = clip_model.get_image_features(**inputs).squeeze().cpu().numpy()

        # Text embedding
        txt_emb = text_model.encode(row["combined_text"])

        # Save everything
        embedded_records.append({
            "file_name": row["file_name"],
            "pil_image": row["pil_image"],                  # retain original image
            "combined_text": row["combined_text"],          # retain original text
            "image_embedding": img_emb,
            "text_embedding": txt_emb
        })

    except Exception as e:
        print(f"Failed to embed {row['file_name']}: {e}")
        continue

# === Step 4: Save Output ===
embedded_df = pd.DataFrame(embedded_records)
```

```python
class HistopathBLIPDataset(Dataset):
    def __init__(self, df, processor, instruction="Provide a medical-grade diagnostic description for this histopathological image, considering features like nuclei size, mitotic activity, and tissue arrangement."
, max_input_len=128, max_target_len=512):
        self.df = df
        self.processor = processor
        self.instruction = instruction
        self.max_input_len = max_input_len
        self.max_target_len = max_target_len

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image = row["pil_image"]
        caption = row["combined_text"]

        # === Process inputs (image + instruction prompt)
        inputs = self.processor(
            images=image,
            text=self.instruction,
            return_tensors="pt",
            padding="max_length",
            truncation=True,
            max_length=self.max_input_len
        )

        # === Tokenize caption as target labels
        labels = self.processor.tokenizer(
            caption,
            return_tensors="pt",
            padding="max_length",
            truncation=True,
            max_length=self.max_target_len
        ).input_ids

        labels[labels == self.processor.tokenizer.pad_token_id] = -100  # ignore padding in loss

        return {
            "pixel_values": inputs["pixel_values"].squeeze(0),
            "attention_mask": inputs["attention_mask"].squeeze(0),
            "input_ids": inputs["input_ids"].squeeze(0),  # optional
            "labels": labels.squeeze(0),
            "text_input": self.instruction
        }
```

```python
from datasets import load_dataset
from transformers import Blip2Processor, Blip2ForConditionalGeneration
from PIL import Image
import torch
from tqdm import tqdm
import pandas as pd

# === Load BLIP-2 fine-tuned model ===
device = "cuda" if torch.cuda.is_available() else "cpu"

model = Blip2ForConditionalGeneration.from_pretrained("/content/drive/MyDrive/patches_captions/blip2-finetuned-histopath-captiongen").to(device)
processor = Blip2Processor.from_pretrained("/content/drive/MyDrive/patches_captions/blip2-finetuned-histopath-captiongen")

# === Load PCam dataset (streaming) ===
dataset = load_dataset("1aurent/PatchCamelyon", split="train", streaming=True)

# === Caption generation function ===
def generate_caption(image: Image.Image, instruction: str = "Describe this histopathology image like a medical expert."):
    inputs = processor(images=image, text=instruction, return_tensors="pt").to(device)

    with torch.no_grad():
        generated_ids = model.generate(**inputs, max_new_tokens=50)

    caption = processor.tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()
    return caption

# === Collect all fields including image + caption ===
records = []

max_samples = 2000  # change this as needed
print("ğŸ”„ Generating captions and storing in DataFrame...")
for i, example in enumerate(tqdm(dataset, total=max_samples)):
    if i >= max_samples:
        break

    try:
        image = example["image"]
        label = example["label"] if "label" in example else None
        caption = generate_caption(image)

        records.append({
            "id": i,
            "label": label,
            "generated_caption": caption,
            "image": image  # note: PIL Image, not serialized
        })

    except Exception as e:
        print(f"âš ï¸ Skipped sample {i} due to error: {e}")
        continue

# === Create DataFrame ===
df = pd.DataFrame(records)

# === Save or use further ===
df.to_pickle("pcam_with_captions.pkl")  # PIL objects are preserved
print("âœ… DataFrame saved to 'pcam_with_captions.pkl'")
```

```python
import pandas as pd

def select_samples_from_df(df, class_column='label', sample_size=5):
    # Group by the class column and sample from each group
    sampled_df = df.groupby(class_column).apply(
        lambda x: x.sample(n=min(len(x), sample_size), random_state=42)
    ).reset_index(drop=True)

    return sampled_df
```

```python
import random
def select_samples(train_dataset, class_column='label', sample_size=5):
    # Group by class using a dictionary
    class_groups = {key: [] for key in set(record[class_column] for record in train_dataset)}

    for record in train_dataset:
        class_groups[record[class_column]].append(record)

    # Select 5 random samples from each class
    selected_samples = []
    for records in class_groups.values():
        selected_samples.extend(random.sample(records, min(len(records), sample_size)))

    return selected_samples
```
